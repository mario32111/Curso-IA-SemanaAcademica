```python
#Grafico salario/posicion
```

```python

import pandas as pd
import matplotlib.pyplot as plt  # Corregido: pyplot en lugar de pylot
#cargar un conjunto de datos en formato CSV
datos = pd.read_csv("football_mini.csv")

ax= plt.subplot()


plt.plot()
plt.title("Relacion Posicion/Salario de Jugadores de Futbol")
ax.set_xlabel("Posicion")
ax.set_ylabel("Salario")
posicion = datos["Position"]
salario = datos ["Wage"]
plt.bar(posicion, salario)
plt.show()
```

```python
#Grafico de dispersion
```

```python
ax = plt.subplot()
ax.scatter(x = datos["Value"], y = datos["Wage"])
ax.set_xlabel("Salario")
ax.set_xlabel("Valor")
plt.title("Grafico de dispersion Salario/Valor")
plt.show()
```

```python
#Grafico de dispersion pero con plot
```

```python

ax = plt.subplot()
ax.plot(datos["Value"],datos["Wage"], marker= "o", ms=4, linestyle="")

ax.set_xlabel("Salario")
ax.set_xlabel("Valor")
plt.title("Grafico de dispersion Salario/Valor")
plt.show()
```
